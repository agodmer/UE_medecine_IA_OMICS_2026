## 🛠 Qu'est-ce que Google Colab ?
Google Colab est un environnement de développement en ligne qui permet d'écrire et d'exécuter du code **Python** directement dans votre navigateur, sans aucune installation sur votre ordinateur. C'est l'équivalent de "Google Docs" mais pour le code.

### 1. Structure d'un Notebook
Un notebook est composé de **cellules** :
- **Cellules de texte** (comme celle-ci) : Pour les explications et les consignes.
- **Cellules de code** (fond gris) : Pour exécuter du code Python.

### 4. Erreurs courantes
- **Oubli de parenthèse** : `print("Hello"` au lieu de `print("Hello")`.
- **Variable non définie** : Vous essayez d'utiliser une variable avant d'avoir exécuté la cellule qui la crée.
  - *Solution :* Allez dans **Exécution > Tout exécuter**.
